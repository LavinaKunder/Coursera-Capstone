```python
Canada_Postcodes = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
Postcode_Table = Canada_Postcodes[0]

print(Postcode_Table.head(6))



```

```python
df = pd.DataFrame(Postcode_Table)
```

```python
df2 = df[df.Borough != 'Not assigned']
```

```python
df3 = df2.groupby(["Postcode","Borough"],sort= False).agg(','.join)
df3.reset_index(inplace=True)
```

```python
df4 = df3.replace("Not assigned","Queen's Park")
```

```python
df_latlon = pd.read_csv("https://cocl.us/Geospatial_data")
df_latlon.columns = ['Postcode','Latitude','Longitude']
df_geo = pd.merge(df4,df_latlon,on=['Postcode'],how='inner')

df_tor = df_geo[['Borough','Neighbourhood','Postcode','Latitude','Longitude']].copy()

df_tor.head()
```

Geographical Coordinates of Toronto

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude_toronto = location.latitude
longitude_toronto = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))
```

Map of Toronto


```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Borough'], df_tor['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
